## Importing required library

```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression 
```

## Loading the data

```python
df1 = pd.read_csv('Breast_Cancer_Data.csv')
df1.head(10)
```

## Data Preprocessing

```python
df1.diagnosis = df1.diagnosis.replace({'M': 1, 'B': 0})
df1.head(10)
```

```python
df1.shape
```

```python
df1.columns
```

```python
df2 = df1.drop(['Unnamed: 32'],axis='columns') # Dropping unnecessary columns
df2.shape
```

```python
df2.isnull().sum() # Checking for null values in our data
```

## Splitting dataset for training and testing

```python
X = df2.drop(['diagnosis'],axis='columns')
X.head()
```

```python
y = df2.diagnosis
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

# Model Fitting

### 1. Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
pred_lr = lr.predict(X_test)
print(classification_report(y_test, pred_lr))
```

### 2. Random Forest Classifier

```python
rfc = RandomForestClassifier(random_state=0)
rfc.fit(X_train, y_train) 
rfc.score(X_test, y_test)
```

```python
pred_rfc = rfc.predict(X_test)
print(classification_report(y_test, pred_rfc))
```

### 3. Decision Tree Classification

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt.score(X_test, y_test)
```

```python
pred_dt = dt.predict(X_test)
print(classification_report(y_test, pred_dt))
```
